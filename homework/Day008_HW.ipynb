### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料？ => r.text')
print('資料型態是什麼？ =>', type(r.text))
```

```python
html_doc = r.text
soup = BeautifulSoup(html_doc, "html5lib")
print('為什麼要使用 BeatifulSoup 處理？ => ', '用 BeatifulSoup 處理之後可以擷取出我們想要的部分')
print('處理後的型態是什麼？ => ', type(soup))
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？
